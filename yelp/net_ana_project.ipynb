```python
df_b = pd.read_csv("datasets/business.csv")
df_b.drop('Unnamed: 0',inplace=True, axis=1)
df_b
```

### Merging review dataset and business dataset

```python
merge=df_b.merge(df_r,on='business_id',how='inner')
merge
```

```python
merge['state'].unique()
```

```python
state1=["CA"]
CA=merge.loc[merge['state'].isin(state1)]
CA
```

a) **review_count** is the column which has total number reviews for each business 

b) **R selected is 201** , so selecting the data with review count more than 201

```python
CA_r201=CA.loc[(CA['review_count'] >= 201)]
CA_r201
```
