```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.star_graph(99) 
# illustrate graph
fig, ax = plt.subplots(figsize=(20,20))
nx.draw(G,node_color = 'green',node_size = 1000,with_labels = True, ax=ax) 
```

```python
a = random.sample(range(0, 99), 10)
b = random.sample(range(0, 99), 10)
```

```python
edges=110
for i in range(0,10):
    G.add_edge(a[i],b[i])
    edges-=1
#nx.draw(G, node_color = 'green',node_size = 100)
fig, ax = plt.subplots(figsize=(20,20))
nx.draw(G,node_color = 'green',node_size = 1000,with_labels = True, ax=ax) 
```

```python
# extracting edges list for the graph
B=list(G.edges)
print(B)
```

```python
# creating a list of lists with zeros same as the size of nodes
adj=[]
for j in range(100):
    adj.append([])
    for k in range(100):
        adj[j].append(0)
```

```python
# The adjacency list can be formed as
for m in B:
    adj[m[0]][m[1]]=1
    adj[m[1]][m[0]]=1
```

```python
A = np.array(adj)
shape = (100,100)
print("ADJACENCY MATRIX IS: \n")
print(A.reshape(shape))
```

```python
D1 = np.zeros([A.shape[0], A.shape[0]])       # creating a zero matrix as the size of adjacency matrix
z, nz = A.nonzero()                          #Return the indices of the elements that are non-zero
for r, c in zip(z, nz):
    D1[r, r] += 1                             # Incrementing the diagonal elements with 1 for each 1 in the row and column
```

```python
D=D1.astype(int)
```

```python
print("DIAGONAL MATRIX IS: \n")
print(D)
```

```python
L=D-A
print("LAPLACIAN MATRIX IS: \n")
print(L)
```
