```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
#G = nx.MultiGraph()
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(5, 1)
G.add_edge(1, 6)
G.add_edge(2, 7)
G.add_edge(3, 8)
G.add_edge(4, 9)
G.add_edge(5, 10)
G.add_edge(6, 8)
G.add_edge(6, 9)
G.add_edge(7, 9)
G.add_edge(7, 10)
G.add_edge(8, 10)

#explicitly set positions
pos = {1: (4, 8), 2: (10, 4), 3: (8, 0), 4: (0, 0), 5: (-2, 4), 6: (4, 6), 7: (8, 4), 8:(6, 2),9:(2,2),10:(0,4)}

options = {
    "font_size": 36,
    "node_size": 2000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 3,
    "width": 3,
}
fig = nx.draw(G, pos, **options, with_labels = True)
```

```python
deg_centrality = nx.degree_centrality(G)
print('Degree Centrality of the Graph G is : \n', deg_centrality)
print('')
print('Degree centrality of Vertex 1 is :\n', deg_centrality[1])
```

```python
b=nx.betweenness_centrality(G)
b
```
